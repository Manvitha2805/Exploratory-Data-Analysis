```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("task1_cleaned_engineered _dataset.csv")
```

```python
df.head()
```

```python
df['Status']=df['Status'].str.strip().str.title()
df['ship-service-level']=df['ship-service-level'].str.strip().str.title()
df['Category']=df['Category'].str.strip().str.title()
df['Size']=df['Size'].str.strip().str.title()
df['ship-state']=df['ship-state'].str.strip().str.title()
df['ship-city']=df['ship-city'].str.strip().str.title()
df['Day Name']=df['Day Name'].str.strip().str.title()
```

```python
df['ship-state']=df['ship-state'].replace({
    'New Delhi': 'Delhi',
    'Orissa':'Odisha',
    'Puducherry':'Pondicherry',
    'RJ':'Rajasthan',
    'Rj':'Rajasthan',
    'Rajsthan':'Rajasthan',
    'Rajshthan':'Rajasthan',
    'Pb':'Punjab',
    'Punjab/Mohali/Zirakpur':'Punjab',
    'Ar':'Arunachal Pradesh'
})
df=df[df['ship-state']!='Nl']
```

```python
df['ship-city']=df['ship-city'].str.strip().str.title()
```

Key Performance Indicators

Proposed Indicators:


The Following kPI's are proposed to evaluate business performance.
*   Total Revenue
*   Total Orders
*   Average Orer Value
*   Total Quantity Sold
*   Top Performing Category

```python
valid_orders=df[df['Status']=='Shipped - Delivered To Buyer']
total_revenue=valid_orders['Total_Revenue'].sum()
total_orders=valid_orders['Order ID'].nunique()
average_order_value=total_revenue/total_orders
total_quantity_sold=valid_orders['Qty'].sum()
top_category=valid_orders.groupby('Category')['Total_Revenue'].sum().idxmax()
print("Total Revenue:",total_revenue)
print("Total Orders:",total_orders)
print("Average Order Value:",average_order_value)
print("Total Quantity Sold:",total_quantity_sold)
print("Top Performing Category:",top_category)
```
