```python
import pandas as pd
import sqlite3
```

```python
df=pd.read_csv("DataSet")
```

```python
conn=sqlite3.connect(':memory:')
df.to_sql('data',conn, index=False)
```

SQL ANALYSIS

Total Revenue

```python
query="""SELECT SUM(Total_Revenue) AS Total_Revenue FROM data"""
pd.read_sql(query,conn)
```

Revenue by Category


```python
query="""SELECT Category, SUM(Total_Revenue) AS Total_Revenue FROM data  GROUP BY Category ORDER BY Total_Revenue DESC"""
pd.read_sql(query,conn)
```

Monthly Revenue

```python
query="""SELECT Month, SUM(Total_Revenue) AS Total_Revenue FROM data GROUP BY Month ORDER BY Month"""
pd.read_sql(query,conn)
```

Top States by Revenue

```python
query="""SELECT "ship-state", SUM(Total_Revenue) AS Revenue FROM data WHERE Status='Shipped - Delivered To Buyer' GROUP BY "ship-state" ORDER BY Revenue DESC Limit 10"""
pd.read_sql(query,conn)
```

Total Orders per Category

```python
query="""SELECT Category, COUNT(*) AS Total_Orders FROM data GROUP BY Category ORDER BY Total_Orders DESC"""
pd.read_sql(query,conn)
```

Total Quantity Sold by Category

```python
query="""SELECT Category, SUM(Qty) AS Total_Quantity FROM data GROUP BY Category ORDER BY Total_Quantity DESC"""
pd.read_sql(query,conn)
```

Order by Status

```python
query="""SELECT Status, COUNT(*) AS Total_Orders FROM data GROUP BY Status ORDER BY Total_Orders DESC"""
pd.read_sql(query,conn)
```

Average Order Value

```python
import pandas as pd
query="""SELECT AVG(order_total) AS Average_Order_Value FROM ( SELECT "Order ID", SUM(Total_Revenue) AS order_total FROM data WHERE Status='Shipped - Delivered To Buyer' GROUP BY "Order ID")"""
pd.read_sql(query,conn)
```
